```python
from transformers import pipeline

# Load the text generation pipeline with GPT-2
generator = pipeline('text-generation', model='gpt2')

def generate_story(prompt):
    """
    Generates a story based on the user input prompt using GPT-2.
    """
    story = generator(prompt, max_length=200, temperature=0.7, num_return_sequences=1)
    return story[0]['generated_text']

def main():
    print("Welcome to the Free Story Generator! 🚀")
    user_prompt = input("Enter your story prompt: ")
    if user_prompt.strip():
        story = generate_story(user_prompt)
        print("\nHere's your generated story:\n")
        print(story)
    else:
        print("Please enter a valid prompt.")

if __name__ == "__main__":
    main()
```
